```python
# Визуализация 2D векторов
def plot_2d_vectors(*vectors, labels=None, colors=None):
    """Функция для визуализации 2D векторов"""
    fig, ax = plt.subplots(figsize=(8, 8))

    if colors is None:
        colors = ['red', 'blue', 'green', 'orange', 'purple']

    if labels is None:
        labels = [f'v{i+1}' for i in range(len(vectors))]

    for i, vector in enumerate(vectors):
        ax.arrow(0, 0, vector[0], vector[1],
                head_width=0.2, head_length=0.2,
                fc=colors[i % len(colors)], ec=colors[i % len(colors)],
                linewidth=2, label=labels[i])

    # Настройка осей
    max_val = max([max(abs(v)) for v in vectors]) + 1
    ax.set_xlim(-max_val, max_val)
    ax.set_ylim(-max_val, max_val)
    ax.grid(True, alpha=0.3)
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.legend()
    ax.set_title('Визуализация векторов')
    plt.show()

# Визуализируем векторы a и b
plot_2d_vectors(a, b, labels=['a = [3, 4]', 'b = [1, 2]'])
```

### Задание 3.2: Косинусная мера сходства для текстов

```python
# Простое представление текстов в виде векторов (bag of words)
# Словарь: ["кот", "собака", "компьютер", "алгоритм", "учиться", "дом", "машина"]

# Документы представлены как векторы частот слов
doc1 = np.array([3, 1, 0, 0, 1, 2, 0])  # "кот кот кот собака учиться дом дом"
doc2 = np.array([2, 2, 0, 0, 1, 1, 0])  # "кот кот собака собака учиться дом"
doc3 = np.array([0, 0, 3, 2, 1, 0, 1])  # "компьютер компьютер компьютер алгоритм алгоритм учиться машина"
doc4 = np.array([1, 1, 1, 1, 2, 0, 1])  # "кот собака компьютер алгоритм учиться учиться машина"

documents = [doc1, doc2, doc3, doc4]
doc_names = ["Документ 1 (животные)", "Документ 2 (животные)", "Документ 3 (технологии)", "Документ 4 (смешанный)"]

print("Векторные представления документов:")
for i, (doc, name) in enumerate(zip(documents, doc_names)):
    print(f"{name}: {doc}")

# Функция для вычисления косинусной близости
def cosine_similarity(vec1, vec2):
    """Вычисляет косинуснцю близость между двумя векторами"""
    # TODO: Реализуйте формулу
    # cos(θ) = (a · b) / (||a|| * ||b||)
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)

    return dot_product / (norm_vec1 * norm_vec2)

# Вычисление матрицы сходства
n_docs = len(documents)
similarity_matrix = np.zeros((n_docs, n_docs))

for i in range(n_docs):
    for j in range(n_docs):
        similarity_matrix[i, j] = cosine_similarity(documents[i], documents[j])

print("\nМатрица косинусного сходства:")
print(similarity_matrix)

# Визуализация матрицы сходства
plt.figure(figsize=(8, 6))
plt.imshow(similarity_matrix, cmap='Blues', interpolation='nearest')
plt.colorbar(label='Косинусное сходство')
plt.xticks(range(n_docs), [f'Док {i+1}' for i in range(n_docs)])
plt.yticks(range(n_docs), [f'Док {i+1}' for i in range(n_docs)])
plt.title('Матрица косинусного сходства документов')

# Добавление значений в ячейки
for i in range(n_docs):
    for j in range(n_docs):
        plt.text(j, i, f'{similarity_matrix[i, j]:.2f}',
                ha='center', va='center', fontweight='bold')

plt.show()
```

### Задание 3.3: Простая линейная регрессия с матрицами

```python
# Простой пример линейной регрессии
# Задача: предсказать цену дома по его площади

# Данные: площадь дома (кв.м) и цена (млн руб.)
areas = np.array([50, 60, 70, 80, 90, 100, 110, 120])  # площадь
prices = np.array([3.2, 3.8, 4.1, 4.7, 5.2, 5.8, 6.1, 6.9])  # цена

# Создание матрицы признаков X (добавляем столбец единиц для свободного члена)
X = np.column_stack([np.ones(len(areas)), areas])
y = prices

print("Матрица признаков X (первые 5 строк):")
print(X[:5])
print(f"\nВектор целевых значений y: {y}")

# Решение методом наименьших квадратов: w = (X^T * X)^(-1) * X^T * y
# TODO: Вычислите веса w используя формулу выше
X_T = X.T  # транспонированная матрица X
XTX = X_T @ X  # произведение X^T и X
XTX_inv = np.linalg.inv(XTX)  # обратная матрица
XTy = X_T @ y  # произведение X^T и y
w = XTX_inv @ XTy # финальное вычисление весов

print(f"\nВеса модели: w = {w}")
print(f"Свободный член (intercept): {w[0]:.3f}")
print(f"Коэффициент при площади: {w[1]:.3f}")

# Предсказания
predictions = X @ w

# Визуализация
plt.figure(figsize=(10, 6))
plt.scatter(areas, prices, color='blue', label='Реальные данные', s=50)
plt.plot(areas, predictions, color='red', label='Линейная регрессия', linewidth=2)
plt.xlabel('Площадь дома (кв.м)')
plt.ylabel('Цена (млн руб.)')
plt.title('Линейная регрессия: Предсказание цены дома')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Предсказание для нового дома
new_area = 95
new_house = np.array([1, new_area])
predicted_price = new_house @ w
print(f"\nПредсказанная цена дома площадью {new_area} кв.м: {predicted_price:.2f} млн руб.")
```

# Задание 1: Евклидово расстояние

```python
def euclidean_distance(vec1, vec2):
    """Вычисляет евклидово расстояние между двумя векторами"""
    vec1 = np.array(vec1)
    vec2 = np.array(vec2)
    if vec1.shape != vec2.shape:
        raise ValueError("Векторы должны быть одинаковой размерности")
    return np.sqrt(np.sum((vec1 - vec2) ** 2))
```

# Задание 2: Нормализация вектора

```python
def normalize_vector(vec):
    """Нормализует вектор (приводит к единичной длине)"""
    vec = np.array(vec, dtype=float)
    norm = np.linalg.norm(vec)
    if norm == 0:
        return vec
    return vec / norm
```

# Задание 3: Матрица поворота

```python
def rotation_matrix_2d(angle_deg):
    """
    Создаёт матрицу поворота на угол angle_deg (в градусах)
    """
    angle_rad = np.radians(angle_deg)
    cos_a = np.cos(angle_rad)
    sin_a = np.sin(angle_rad)
    return np.array([
        [cos_a, -sin_a],
        [sin_a, cos_a]
    ])

def rotation_matrix_3d(angle_deg, axis='z'):
    """
    Создаёт матрицу поворота в 3D вокруг оси X, Y или Z
    """
    angle_rad = np.radians(angle_deg)
    cos_a = np.cos(angle_rad)
    sin_a = np.sin(angle_rad)

    if axis == 'x':
        return np.array([
            [1, 0, 0],
            [0, cos_a, -sin_a],
            [0, sin_a, cos_a]
        ])
    elif axis == 'y':
        return np.array([
            [cos_a, 0, sin_a],
            [0, 1, 0],
            [-sin_a, 0, cos_a]
        ])
    elif axis == 'z':
        return np.array([
            [cos_a, -sin_a, 0],
            [sin_a, cos_a, 0],
            [0, 0, 1]
        ])
    else:
        raise ValueError("Ось должна быть 'x', 'y' или 'z'")

# Тестирование матрицы поворота
# 2D поворот
vec = np.array([1, 0])
angle = 45
R = rotation_matrix_2d(angle)
rotated_vec = R @ vec

print(f"\n2D Поворот на {angle}°:")
print(f"  Исходный вектор: {vec}")
print(f"  Матрица поворота:\n{R}")
print(f"  Повёрнутый вектор: {rotated_vec}")
print(f"  Длина вектора: {np.linalg.norm(rotated_vec):.3f} (сохранилась)")

# Визуализация поворота
fig, ax = plt.subplots(figsize=(6, 6))
angles = [0, 30, 60, 90, 120, 150, 180]
colors = plt.cm.viridis(np.linspace(0, 1, len(angles)))

for angle, color in zip(angles, colors):
    R = rotation_matrix_2d(angle)
    rotated = R @ vec
    ax.quiver(0, 0, rotated[0], rotated[1], angles='xy', scale_units='xy',
              scale=1, color=color, alpha=0.7, width=0.015,
              label=f'{angle}°')

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.grid(True, alpha=0.3)
ax.set_title('Поворот вектора (1, 0) на разные углы')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend(loc='upper left')
ax.axhline(y=0, color='black', linewidth=0.5)
ax.axvline(x=0, color='black', linewidth=0.5)
ax.set_aspect('equal')
plt.show()
```

# Задание 4: Косинусная близость с шумом

```python
def cosine_similarity(vec1, vec2):
    """Вычисляет косинусное сходство между двумя векторами"""
    vec1 = np.array(vec1)
    vec2 = np.array(vec2)
    dot = np.dot(vec1, vec2)
    norm1 = np.linalg.norm(vec1)
    norm2 = np.linalg.norm(vec2)
    if norm1 == 0 or norm2 == 0:
        return 0
    return dot / (norm1 * norm2)

def add_noise(vector, noise_level=0.1):
    """Добавляет гауссов шум к вектору"""
    noise = np.random.normal(0, noise_level, size=len(vector))
    return vector + noise

# Генерируем случайные векторы
np.random.seed(42)
n_vectors = 50
dim = 10
noise_levels = np.linspace(0, 1.0, 11)

# Создаём базовый вектор
base_vector = np.random.randn(dim)
base_vector = normalize_vector(base_vector)

# Создаём набор векторов
vectors = []
for _ in range(n_vectors):
    v = np.random.randn(dim)
    vectors.append(normalize_vector(v))

# Исследуем влияние шума
print("\nВлияние уровня шума на косинусное сходство:")
print("-" * 50)
print(f"{'Уровень шума':<15} {'Среднее сходство':<20} {'Станд. отклонение':<20}")
print("-" * 50)

results = []
for noise_level in noise_levels:
    similarities = []
    for v in vectors:
        noisy_v = add_noise(v, noise_level)
        noisy_v = normalize_vector(noisy_v)
        sim = cosine_similarity(v, noisy_v)
        similarities.append(sim)

    mean_sim = np.mean(similarities)
    std_sim = np.std(similarities)
    results.append((noise_level, mean_sim, std_sim))
    print(f"{noise_level:<15.2f} {mean_sim:<20.4f} {std_sim:<20.4f}")

# Визуализация
fig, ax = plt.subplots(figsize=(10, 6))
noise_vals, mean_vals, std_vals = zip(*results)

ax.errorbar(noise_vals, mean_vals, yerr=std_vals,
            fmt='o-', capsize=5, capthick=2,
            color='blue', ecolor='red',
            label='Косинусное сходство с шумом')

ax.axhline(y=1.0, color='green', linestyle='--', alpha=0.5, label='Идеальное сходство (без шума)')
ax.set_xlabel('Уровень шума')
ax.set_ylabel('Косинусное сходство')
ax.set_title('Зависимость косинусного сходства от уровня шума')
ax.grid(True, alpha=0.3)
ax.legend()
plt.show()
```

# Задание 5: Исследование различных расстояний

```python
from scipy.spatial.distance import euclidean, cityblock, chebyshev, cosine, correlation, minkowski

# Создаём тестовые данные
np.random.seed(42)
X = np.random.randn(10, 5)
Y = np.random.randn(10, 5)

print("\nСравнение различных расстояний между векторами X[0] и Y[0]:")
print("-" * 60)

vec1 = X[0]
vec2 = Y[0]

distances = {
    'Евклидово (L2)': euclidean(vec1, vec2),
    'Манхэттенское (L1)': cityblock(vec1, vec2),
    'Чебышёва (L∞)': chebyshev(vec1, vec2),
    'Косинусное расстояние': cosine(vec1, vec2),
    'Корреляционное': correlation(vec1, vec2),
    'Минковского (p=3)': minkowski(vec1, vec2, p=3)
}

for name, value in distances.items():
    print(f"  {name:<25}: {value:.4f}")

print("Объяснение расстояний:")
print("""
  1. Евклидово (L2) — 'прямая' дистанция, чувствительна к выбросам
  2. Манхэттенское (L1) — сумма абсолютных разностей, устойчивее к выбросам
  3. Чебышёва (L∞) — максимальная разность по любой координате
  4. Косинусное — измеряет угол между векторами (1 - косинусное сходство)
  5. Корреляционное — 1 - корреляция Пирсона
  6. Минковского — обобщение L1 и L2 (параметр p)
""")

# Демонстрация на реальных данных
print("\n" + "-"*60)
print("Пример использования расстояний для кластеризации (k-means):")
print("-"*60)

# Создаём кластеры точек
np.random.seed(42)
cluster1 = np.random.randn(30, 2) + [0, 0]
cluster2 = np.random.randn(30, 2) + [5, 5]
data = np.vstack([cluster1, cluster2])

# Визуализация
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Исходные данные
ax1 = axes[0]
ax1.scatter(cluster1[:, 0], cluster1[:, 1], c='blue', alpha=0.6, label='Кластер 1')
ax1.scatter(cluster2[:, 0], cluster2[:, 1], c='red', alpha=0.6, label='Кластер 2')
ax1.set_title('Исходные данные с двумя кластерами')
ax1.set_xlabel('Признак 1')
ax1.set_ylabel('Признак 2')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Сравнение расстояний
ax2 = axes[1]
center = np.array([2.5, 2.5])
distances_to_center = []

for point in data:
    d_euclidean = euclidean(point, center)
    d_manhattan = cityblock(point, center)
    distances_to_center.append((d_euclidean, d_manhattan))

d_euclidean_list, d_manhattan_list = zip(*distances_to_center)
ax2.scatter(d_euclidean_list, d_manhattan_list, c=['blue']*30 + ['red']*30, alpha=0.6)
ax2.set_xlabel('Евклидово расстояние до центра')
ax2.set_ylabel('Манхэттенское расстояние до центра')
ax2.set_title('Сравнение расстояний от центра (2.5, 2.5)')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
